# Practical 6: Constrained Optimization and Lagrange Multipliers
    Understanding Constrained Optimization using Lagrange Multipliers

```python
import sympy as sp
```

```python
x, y, lam = sp.symbols('x y lam') 

#x,y  = optimization variables ,lam  =  Lagrange multiplier    
```

```python
f = x**2 + y**2

constraint = x + y - 10  #define objective
```

```python
L = f + lam * constraint

print("Lagrangian Function:")
print(L)
```

```python

#calculate partial derivative
dL_dx = sp.diff(L, x)

dL_dy = sp.diff(L, y)

dL_dlam = sp.diff(L, lam)

print(dL_dx)
print(dL_dy)
print(dL_dlam)
```

```python
solution = sp.solve(
    [dL_dx, dL_dy, dL_dlam],
    (x, y, lam)
)

print(solution)
```
